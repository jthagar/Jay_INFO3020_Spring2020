```python
# methods for question 2 of 10-2
def delete_student():
    stuID = input("input student ID for deletion: ")
    sql = "delete from student_table where stuid = %s"
    stu_cursor.execute(sql,stuID)
    student.commit()
    
    
def change_student_name():
    oldie = input("input student name you wish to change: ")
    newie = input("input the new name: ")
    sql = "update student_table set stuname=%s where stuname=%s"
    vals = (oldie, newie)
    
    stu_cursor.execute(sql, vals)
    student.commit()
    
def get_num_majors():
    track = str(input("type major to fetch students from: "))
    sql = "select * from student_table where major=%s"
    stu_cursor.execute(sql,track)
    
    trackers = stu_cursor.fetchall()
    print("students fetched: \n")
    for kid in trackers:
        print(str(kid[0]) +"\t"+kid[1]+"\t"+kid[2]+"\n")

import time
import random as rand 
def get_rand_student():
    rand.seed(time.time())
    
    query = "select count(*) from student_table"
    size = stu_cursor.execute(query)
    row = int(rand.randint(0,size - 1)) #the minus 1 is for the list being indexed at zero
    query = "select * from student_table"
    
    stu_cursor.execute(query)
    all_stu = stu_cursor.fetchall()
    random_stu = all_stu[row]
    
    print("random student chosen from all students: \n")
    print( str(random_stu[0]) + "\t" + random_stu[1] )



get_num_majors()
get_rand_student()
change_student_name()
delete_student()

    
student.close()
```
